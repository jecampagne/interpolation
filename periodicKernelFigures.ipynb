# Kernel approx of wrapped/periodic Sinc

```python
u=np.linspace(-2.5,2.5,10_000)
for n in [5,10,100]:
  plot(u,np.abs(np.sinc(n*u)/np.sinc(u)),label=f"N={n}")
xlabel("x")
legend()
title(r"$|sinc(Nx)/sinc(x)|$");
```

```python
N=16
```

```python
figure(figsize=(6,4))
u=np.linspace(-2.5,2.5,10_000)
plot(u,np.sinc(N*u)/np.sinc(u),label=r"$Sinc(N\nu)/Sinc(\nu)$")
#plot(u,rlanczosWrapped(u,N,m=3),label="Lanczos 3",ls='--')
#plot(u,rlanczosWrapped(u,N,m=5),label="Lanczos 5",ls='--')
#plot(u,rlanczosRenormWrapped(u,N,m=3),label="Lanczos 3 renorm",ls='--')
plot(u,rlanczosRenormWrapped(u,N,m=10),label="Lanczos 5 renorm",ls='--')
plot(u,rcubicWrapped(u,N),label="Cubic",ls='--')
plot(u,rquinticBGWrapped(u,N),label="Quintic BG",ls='--')
plot(u,rquinticJEWrapped(u,N),label="Quintic JE",ls='--')
legend(loc='center right', bbox_to_anchor=(1.45, 0.5),fontsize=12);
title(f"Periodic kernels with N={N}");
xlabel(r"$\nu$")
grid()
```

```python
figure(figsize=(6,4))
u=np.linspace(-0.5,0.5,10_000)
plot(u,np.sinc(N*u)/np.sinc(u),label=r"$Sinc(N\nu)/Sinc(\nu)$")
#plot(u,rlanczosWrapped(u,N,m=3),label="Lanczos 3",ls='--')
#plot(u,rlanczosWrapped(u,N,m=5),label="Lanczos 5",ls='--')
#plot(u,rlanczosRenormWrapped(u,N,m=3),label="Lanczos 3 renorm",ls='--')
plot(u,rlanczosRenormWrapped(u,N,m=5),label="Lanczos 5 renorm",ls='--')
plot(u,rcubicWrapped(u,N),label="Cubic",ls='--')
plot(u,rquinticBGWrapped(u,N),label="Quintic BG",ls='--')
plot(u,rquinticJEWrapped(u,N),label="Quintic JE",ls='--')
legend(loc='center right', bbox_to_anchor=(1.45, 0.5),fontsize=12);
grid()
xlabel(r"$\nu$")
title(f"Periodic kernels with N={N}");
```
